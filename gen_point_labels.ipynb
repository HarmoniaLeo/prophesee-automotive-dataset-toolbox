```python
import numpy as np
from src.io.psee_loader import PSEELoader

name="/Volumes/SN550_2/Large_Automotive_Detection_Dataset/train/moorea_2019-06-26_test_02_000_976500000_1036500000_"
```

```python
video = PSEELoader(name+"td.dat")
arr=PSEELoader(name+"bbox.npy")
deltat=20000
```

```python
events = video.load_delta_t(deltat)
boxes = arr.load_delta_t(deltat)
xlow=boxes["x"]
ylow=boxes["y"]
xhigh=boxes["x"]+boxes["w"]
yhigh=boxes["y"]+boxes["h"]
eventsx=np.array([events["x"]]).T
eventsy=np.array([events["y"]]).T
crit1=(eventsx>xlow)
crit2=(eventsx<xhigh)
crit3=(eventsy>ylow)
crit4=(eventsy<yhigh)
crit=(crit1&crit2&crit3&crit4)
print((np.any(crit,axis=1))
print(np.argmax(crit,axis=1)[:5],np.argmax(crit,axis=1).shape)
```

```python
while not video.done:
    events = video.load_delta_t(deltat)
    boxes = arr.load_delta_t(deltat)
    print(len(events),len(boxes))
```

```python
begin_time = 0
sample_time = np.unique(arr["t"])
arr2=None
for end_time in sample_time:
    bbox_list=arr[arr["t"]==end_time]
    video.seek_time(begin_time)
    point_cloud=video.load_delta_t(end_time-begin_time)
    arr2buf=np.zeros((point_cloud.shape[0],2))
    arr2buf[:,0]=arr2buf[:,0]-1
    for i,bbox in enumerate(bbox_list):
        arr2buf[:,1]=np.where((point_cloud["x"]>bbox[1])&(point_cloud["x"]<bbox[1]+bbox[3])&(point_cloud["y"]>bbox[2])&(point_cloud["y"]<bbox[2]+bbox[4])&(arr2buf[:,1]==-1),i,-1)
    arr2buf[:,-1]=arr2buf[:,-1]+end_time
    if arr2 is None:
        arr2=arr2buf
    else:
        arr2=np.vstack((arr2,arr2buf))
    begin_time=end_time
    print(end_time/sample_time[-1],len(arr2buf))
```

```python
print(video.event_count())
print(arr2.shape)
```

```python
np.save(name+"pl.npy",arr2)
```
